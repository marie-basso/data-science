### Removing duplicates, handling missing values, data formatting

```python
print('Number of duplicate rows:', len(df[df.duplicated()]))
```

```python
df_clean = df.drop_duplicates()
print('Number of duplicate rows after cleaning:', len(df_clean[df_clean.duplicated()]))
```

```python
def missing_values(data):
    print('Missing values in each column')    
    miss_val_count = data.isna().sum().sort_values(ascending=False)
    miss_val_percentage = ((miss_val_count / len(data)) * 100).round(2)
    return pd.concat([miss_val_count,miss_val_percentage], axis=1, keys=['count', 'percentage'])
```

```python
missing_values(df_clean)
```

```python
# We don't need these columns
df_clean = df_clean.drop(columns=['county', 'VIN', 'id', 'url', 'image_url'])
```

```python
def useful_rows(data):
    print('The number of fully filled rows:', 
          len(data.dropna()))
    print('Percentage of fully filled rows:', 
          round(len(data.dropna()) / len(data) * 100, 2), '%')
```

```python
useful_rows(df_clean)
```

```python
def unfilled_columns(data, threshold_min, threshold_max):
    # Columns list with missing values
    unfilled_columns = []
    for col in data:
        if threshold_max > (data[col].isna().sum() / len(data))*100 > threshold_min:
            unfilled_columns.append(col)
    return unfilled_columns
```

```python
print('Columns with the minimum number of missing values')
df_clean[unfilled_columns(df_clean, 0, 20)].head(10)
```

```python
def percentage_values(column_name):
    print(f'Percentage of occurrence of each {column_name.name} value in the dataset')
    count = column_name.value_counts(dropna=False)
    percentage = (count / len(column_name)*100).round(2)
    return pd.concat([count,percentage], axis=1, keys=['count', 'percentage'])
```

```python
def column_missing_values(column_name):
    print(f'Number of missing values in the {column_name.name} column: {column_name.isna().sum()}')
```

```python
def column_stat(column_name):
    column_missing_values(column_name)
    print('Number of unique values:', column_name.nunique())
    print(f'{column_name.name} column statistics:')
    return column_name.describe()
```
